### 1. Mistakes
❌ `cols_to_drop = [["Name " + l for l in langs]]`  means a list wrapped in another <br>
✅ `cols_to_drop = (["Name " + l for l in langs])` mean a singular flat list -- which is what we need
<br><br>
❌ heritage_df_rename = heritage_df.rename(cols_to_rename)<br>
✅ heritage_df_rename = heritage_df.rename(columns=cols_to_rename)

```python
# cleaning
langs = ['FR', 'ES', 'RU', 'AR', 'ZH']
# group it for readability, not tuple()
cols_to_drop = (
    # other langs
    ["Name " + l for l in langs] +
    ["Short Description " + l for l in langs] +
    ["Main Image Caption " + l for l in langs] +
    ["Main Video Caption " + l for l in langs] +

    # unwanted: how do we know which? == back to .info()
    # null or unnecessary
    ["Cultural Criteria", "Natural Criteria", "Main Video Caption EN",
     "Date end", "Danger list", "Main Video", "Main Video Author",
     "videos_urls", "UUID", "ID", "Danger", "Criteria", "Secondary dates",
     "Description EN", "Transboundary", "Components", "Main Image Copyright",
     "Area hectares", "Category ID"
    ]
)
# group cols to rename
cols_to_rename = {
    "Name EN": "Name",
    "Short Description EN": "Description",
    "Justification EN": "Justification",
    "States Names": "Countries",
    "Date inscribed": "Date",
    "Main Image": "Initial Image",
    "Main Image Author": "Image Author",
    "Main Image Caption EN": "Image Caption",
}

# process dropping and renaming
heritage_df = heritage.drop(cols_to_drop, axis=1, errors='ignore')
heritage_df_rename = heritage_df.rename(columns=cols_to_rename)

heritage_df_rename.info()
print("Validate dropping: ", heritage_df.columns.to_list())
print("Validate renaming: ", heritage_df_rename.columns.to_list())
heritage_df_rename.head(15)
```

#### Explanation
1. `.str.split(",")` splits the string at the comma <br>  →  ["34.3964166667", "64.5158888889"]<br><br>
2. `.str.split(",", expand=True)` get a list in one column<br>
    | expand=False       | expand=True | expand=True |
    |--------------------|-------------|-------------|
    | 0                  | 0           | 1           |
    | ["34.39", "64.51"] | "34.39"     | "64.51"     |
    | ["12.34", "56.78"] | "12.34"     | "56.78"     |
<br>
3. `.astype(float)` converts string into actual numbers
            <br> == `"34.3964166667"  →  34.3964166667`<br><br>
4. `heritage_df_rename[['Latitude', 'Longitude']] = ...` → a, b = [1, 2]
<br><br>

| Step                                 | Col 0               | Col 1                |
|--------------------------------------|---------------------|----------------------|
| Raw string                           | `"34.3964,64.5158"` | —                    |
| After `.str.split(',', expand=True)` | `"34.3964"`         | `"64.5158"`          |
| After `.astype(float)` → renamed     | `Latitude: 34.3964` | `Longitude: 64.5158` |

```python
# separate into long and lat by a comma
heritage_df_rename[['Latitude', 'Longitude']] = (
    heritage_df_rename['Coordinates'].str.split(",", expand=True).astype(float)
)
heritage_df_rename[['Coordinates', 'Latitude', 'Longitude']].head(20)
```

#### Natures of Pandas: return a new DataFrame
- by the action `heritage_df_rename = heritage_df_rename.explode('Countries')`, assigns back to the original dataframe
- `.explode()` takes a column with lists and create a new row for each item

```python
heritage_df_rename.columns.to_list()
```

```python
# change the index number
heritage_df_rename.index = range(1, len(heritage_df_rename) + 1)
print("Size of data BEFORE exploding by column 'Countries': ", heritage_df_rename.shape)

# # .explode() expected a list, but we had strings
# # split to string by "," to a list
heritage_df_rename['Countries'] = heritage_df_rename['Countries'].str.split(",")
heritage_df_rename['ISO Codes'] = heritage_df_rename['ISO Codes'].str.split(",")
heritage_df_exploded = heritage_df_rename.explode(['Countries', 'ISO Codes'])

print("Size of data AFTER exploding by column 'Countries': ", heritage_df_exploded.shape, "\n")
heritage_df_exploded.head(5)
```

```python
# categorise region into continents
heritage_df_exploded['Continent'] = coco.convert(
    heritage_df_exploded['Countries'], to='continent'
)
heritage_df_exploded['Continent'].value_counts()
```

```python
# add south America
south_america = [
    'AR', 'BO', 'BR', 'CL', 'CO', 'EC',
    'GY', 'PY', 'PE', 'SR', 'UY', 'VE'
]

heritage_df_exploded['Continent'] = [
    'South America' if code in south_america
    else 'North America' if continent == 'America'
    else continent
    for code, continent in zip(heritage_df_exploded['ISO Codes'], heritage_df_exploded['Continent'])
]
heritage_df_exploded['Continent'].value_counts()
```

```python
# check overview stats
print("Total sites: ", heritage_df_exploded['Name'].nunique())
print("Total countries: ", heritage_df_exploded['Countries'].nunique())
print("Year range: ", heritage_df_exploded['Date'].min(), "-", heritage_df_exploded['Date'].max(), "\n")

# top 10
print("Top 10 Countries by numbers of sites: ")
heritage_df_exploded.groupby('Countries')['Name'].nunique().sort_values(ascending=False).head(10)
```

```python
# inscription over time
heritage_df_exploded.groupby('Date')['Name'].nunique().sort_values()
```

```python
# category
heritage_df_exploded['Category'].value_counts()
heritage_df_exploded.groupby(['Continent', 'Category'])['Name'].nunique().unstack()
```

```python
top10 = (
    heritage_df_exploded
    .groupby('Countries')['Name']
    .nunique()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
top10.columns = ['Country', 'Sites']
top10
```

```python
# clean up long official country names for display
name_fixes = {
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Iran (Islamic Republic of)': 'Iran',
    'Russian Federation': 'Russia',
    'Bolivia (Plurinational State of)': 'Bolivia',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'Korea, Republic of': 'South Korea',
    "Lao People's Democratic Republic": 'Laos',
    'Syrian Arab Republic': 'Syria',
    'Viet Nam': 'Vietnam',
    'Tanzania, United Republic of': 'Tanzania',
    'Congo': 'Republic of Congo',
    'Democratic Republic of the Congo': 'DR Congo',
    "Côte d'Ivoire": 'Ivory Coast',
    'Netherlands (Kingdom of the)': 'Netherlands',
}

heritage_df_exploded['Countries'] = heritage_df_exploded['Countries'].replace(name_fixes)

# check
heritage_df_exploded['Countries'].value_counts().head(15)
```

```python
cols = ['Name', 'Description', 'Category', 'Countries', 'ISO Codes', 
        'Region', 'Continent', 'Date', 'Latitude', 'Longitude']

heritage_df_exploded[cols].to_json(
    'heritage.json', 
    orient='records', 
    indent=2,
    force_ascii=False
)
print("Done!")
```
